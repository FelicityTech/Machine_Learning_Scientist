```python
import pandas as pd
import numpy as np

diabetes_df = pd.read_csv('diabetes_clean.csv')
print(diabetes_df.head())
```

```python
X = diabetes_df.drop('glucose', axis=1).values
y = diabetes_df['glucose'].values
print(type(X), type(y))
```

```python
# MAking predictions from a single feature
x_bmi = X[:,3]  # BMI is the 4th column in the dataset
print(X.shape, x_bmi.shape)
```

```python
x_bmi = x_bmi.reshape(-1, 1)  # Reshape to 2D array for sklearn
print(x_bmi.shape)
```

```python
# Plotting glucose vs. Body Mass Index (BMI)
import matplotlib.pyplot as plt
plt.scatter(x_bmi, y)
plt.title('Glucose vs. Body Mass Index (BMI)')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Blood Glucose (mg/dL)')
plt.show()
```

```python
# Fitting a regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_bmi, y)
predictions = reg.predict(x_bmi)
plt.scatter(x_bmi, y)
plt.plot(x_bmi, predictions, color='red')
plt.title('Linear Regression: Glucose vs. BMI')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Blood Glucose (mg/dL)')
plt.show()
```
